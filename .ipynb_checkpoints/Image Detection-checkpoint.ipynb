```python
import mxnet as mx
import gluoncv as gcv
import matplotlib.pyplot as plt
```

```python
# Load Image
image = mx.image.imread('Images/dog_on_a_cycle.jpeg')
```

```python
plt.imshow(image.asnumpy())
```

```python
image , chw_image = gcv.data.transforms.presets.yolo.transform_test(image,short=512)
```

```python
network = gcv.model_zoo.get_model('yolo3_darknet53_coco',pretrained=True)
```

```python
predictions = network(image)
```

```python
for index,array in enumerate(predictions):
    print(index,":",array.shape)
```

```python
predictions = [array[0] for array in predictions]
```

```python
c_indices,c_prob,bounding_box = predictions
```

```python
# Bounding boxes

gcv.utils.viz.plot_bbox(chw_image,bounding_box,c_prob,c_indices,class_names=network.classes)
```
